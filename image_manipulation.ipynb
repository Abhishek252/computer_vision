```python
img = cv2.imread('./img/deer.jpg')
cv2_imshow(img)
```

```python
dst = cv2.fastNlMeansDenoisingColored(img, None, 7, 20, 7, 15)
display = np.hstack((img, dst))
cv2_imshow(display)
```
